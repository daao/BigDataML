```python
import pandas as pd

#Takes about one minute to load
data=pd.read_csv("in/data2.conv.txt",header=None,sep=" ")
data.columns=["Date","Hour","Sensor","Value","Voltage"]
data=data.sort_values(['Date','Hour']).reset_index(drop=True)

data['datetime']=pd.to_datetime(data.Date+' '+data.Hour)
data['relative_datetime']=data['datetime']-data['datetime'][0]
data['seconds']=data['relative_datetime'].dt.total_seconds()

sensorId_type=data.Sensor.str.split("-",expand=True)
sensorId_type.columns=['SensorId','Type']
data['SensorId']=sensorId_type['SensorId'].astype(int)
data['Type']=sensorId_type['Type'].astype(int)

#Drop features not needed for the simulation
data=data.drop(['datetime','relative_datetime','Sensor','Date','Hour','Voltage'],axis=1)

temp=data[(data.SensorId==1) & (data.Type==0)]
temp=temp.reset_index(drop=True)
```

```python
def load_data(split, trainingSet = [], testSet= []):
    for d in data:
        if random.random() < split:
            trainingSet.append([d.Value, d.seconds, d.SensorId, d.Type])
        else:
            testSet.append([d.Value, d.seconds, d.SensorId, d.Type])
```

```python
import math
def euclideanDistance(instance1, instance2, length):
    print(instance1 + " " + instance2)
    distance = pow((instance1[1] - instance2[1]), 2)
    return math.sqrt(distance)
```

```python
import operator
def getNeighbors(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))
    distances.sort()
    neighbors = [distances[x][0] for x in range(k)]
    return neighbors
```

```python
def get_prediction(neighbors):
    votes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][1]
        if response in votes:
            votes[response] += 1
        else:
            votes[response] = 1
    sortedVotes = sorted(votes.items(), reverse=True)
    return sortedVotes[0][0]
```

```python
def mean_squared_error(predictions, values):
    n = len(predictions)
    mse = 0
    for x in range(n):
       mse += (values[x][1] - predictions[x])**2
    return (1/n)*mse
```

```python
training = []
test = []
split = 0.67
load_data(split, training, test)
print('Train set = ' + str(len(training)) + " " + str(training))
print('Test set = ' + str(len(test)))

# generate prediction
predictions = []
k = 3
for x in range(len(test)):
    neighbors = getNeighbors(training, test[x], k)
    response = get_prediction(neighbors)
    predictions.append(response)
    print("Predicted = " + str(response) + ", actual = " + str(test[x][1]))

print("MSE : " + str(mean_squared_error(predictions, test)))
```
